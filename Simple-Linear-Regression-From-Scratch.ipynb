```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# SIMPLE LINEAR REFRESSION

```python
df = pd.DataFrame({"cgpa": [5.5, 6.5, 7.5, 5.8, 6.9, 8.8, 4.9, 6.6, 7.3, 5.9],
                   "pkg": [50000, 60000, 70000, 55000, 75000, 90000, 45000, 62000, 85000, 58000]})

```

```python
df
```

```python
plt.scatter(df["cgpa"],df["pkg"])
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.4,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
x_test
```

```python
lr.predict(x_test.iloc[3].values.reshape(1,1))
```

```python
plt.scatter(df["cgpa"],df["pkg"])
plt.plot(x_test,lr.predict(x_test),color='red')
```

```python
#SLOPE
m=lr.coef_
```

```python
m
```

```python
#y intercept
b=lr.intercept_
```

```python
b
```
